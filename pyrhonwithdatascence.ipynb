```python
import numpy as np

matrix1 = affan.array([[1,2],[3,4]])

matrix2 = affan.array([[5,6],[7,8]])

add = matrix1 + matrix2

matrix_mul = np.dot(matrix1,matrix2)

print(add)

print(matrix_mul)
```

```python
import pandas as pd

data = {
    "Name" : ["affan","suresh","ramesh"],
    "Age" : [23,25,27],
    "City" : ["chennai","ambur","madurai"]
}

df = pd.DataFrame(data)

grped_data= df.groupby('City')['Age'].mean()

print(grped_data)
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y,marker = 'o')
plt.title("basic graph")
plt.xlabel("x-Axis")
plt.ylabel("y-Axis")

plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ('categories-A','categories-B','categories-C')

value = (10,15,25)

plt.bar(categories,value,color="red")
plt.title('basic graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
